```python
from astropy.io import ascii
import pandas as pd
import matplotlib.pyplot as plt

#paper
#https://ui.adsabs.harvard.edu/abs/2021ApJ...922..154M/abstract

#data
#http://www.tauceti.caltech.edu/kunal/gw170817/gw170817_afterglow_data_full.txt
data = ascii.read('gw170817_afterglow_data_full.txt')
data = data.to_pandas()



data["FluxD"] = data["FluxD"].apply(lambda x : float(x.lstrip("<")))

data
```

-----------------------------------------------------------------------------------------------
using only VLA 3GHz data


```python
def plot_data(data, col1 = "T", col2 = "FluxD", col3 ="FluxDErr" ):
    plt.scatter(data[col1], data[col2], cmap="GnBu")
    plt.errorbar(data[col1], data[col2], yerr=data[col3],ecolor="red", linestyle = 'None')
    plt.ylabel("Flux Desnity(micro Jansky)")
    plt.xlabel("Time(days)")
    #plt.plot()
```

```python
VLA_3GHzdata = data.loc[(data["Telescope"]=="VLA") & (data["Freq"]==3.0e9)]
VLA_3GHzdata
```

-----------------------------------------------------------------------------------------------

```python
plot_data(VLA_3GHzdata)
```

```python
plot_data(VLA_3GHzdata.loc[VLA_3GHzdata["T"]<600])
```
